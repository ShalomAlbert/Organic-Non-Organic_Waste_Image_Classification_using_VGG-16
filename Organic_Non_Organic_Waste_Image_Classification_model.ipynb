```python
from tensorflow.keras.models import load_model
model = load_model('/content/drive/MyDrive/Colab Notebooks/Image_classification/classification.h5')
```

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
from tensorflow.keras.preprocessing import image
# ["batteries", "clothes", "e-waste", "glass", "light blubs", "metal", "organic", "paper", "plastic"]
output_class = ["Non-Organic(batteries)", "Non-Organic", "Non-Organic", "Non-Organic", "Non-Organic", "Non-Organic", "Organic", "Organic(paper)", "Non-Organic"]
def waste_prediction(new_image):
  test_image = image.load_img(new_image, target_size = (224,224))
  plt.axis("off")
  plt.imshow(test_image)
  plt.show()

  test_image = image.img_to_array(test_image) / 255
  test_image = np.expand_dims(test_image, axis=0)

  predicted_array = model.predict(test_image)
  predicted_value = output_class[np.argmax(predicted_array)]
  predicted_accuracy = round(np.max(predicted_array) * 100, 2)

  print("Your waste material is ", predicted_value, " with ", predicted_accuracy, " % accuracy")
```
